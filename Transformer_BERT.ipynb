```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis') # an example of pipeline
# pipeline : call a pretrained model (BERT or GPT or ... ) and its tokenizer are downloded
# tokenizer : preprocess the text for the model -> prediction
# pipeline groups all of that together and post-process the predictions to make them readable
```

```python
classifier('I am a :) person')
```

```python
# list of sentences - fed to th emodel as a batch - return a list of dictionaries
results =classifier(["I have a good mother :)", "I really love my mom.","i have a difficult exam","I am sad with hope )"])
results
```

```python
for result in results:
  print(f"label:{result['label']}, score:{round(result['score'],4)}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name="nlptown/bert-base-multilingual-uncased-sentiment"
# This model only exists in PyTorch so we use the 'from_pt' flag to import this model to TensorFlow
model=TFAutoModelForSequenceClassification.from_pretrained(model_name,from_pt=True)
tokenizer=AutoTokenizer.from_pretrained(model_name)
classifier=pipeline('sentiment-analysis', model=model,tokenizer=tokenizer )
```

```python
classifier("I am a good girl") # greater than 3 is positive
```
